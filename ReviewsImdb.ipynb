```python
import pandas as pd
import numpy as np
import re
import lxml
import csv
import matplotlib

from bs4 import BeautifulSoup
from requests import get
```

```python
newimdbdata = pd.read_csv('imdbdatafinal.csv')
```

```python
def number_of_reviews(suburl):
    try:
        subpage = get(suburl)
        subsoup = BeautifulSoup(subpage.content, 'lxml')
        content = subsoup.find(id="main_top")
        titleReviewBar = content.find('div', class_ = 'titleReviewBar')
        titleReviewBarItem = titleReviewBar.find('div', class_ = 'titleReviewBarItem titleReviewbarItemBorder')
        trb = titleReviewBarItem.find('span', class_ = 'subText').text
        semifinal = trb.split('\n')[1]
        final = semifinal.split(' ')[0]
        return final
    except:
        return None
```

```python
numberreviews=[]
for url in newimdbdata['urls']:
    numberreviews.append(number_of_reviews(url))
```

```python
newimdbdata['no. of reviews'] = numberreviews
```

## Read in filtered movied > 500 reviews

```python
filteredimdbdata = pd.read_csv('filteredimdbdata.csv')
```

```python
review_tail = 'reviews?ref_=tt_urv'
```

```python
newurldf = pd.DataFrame({'urls':filteredimdbdata['urls'].copy()})
```

```python
len(newurldf)
```

```python
reviewurls = []
for url in newurldf['urls']:
    complete_url=url+review_tail
    reviewurls.append(complete_url)
```

```python
filteredimdbdata['reviewurls']=reviewurls
```

```python
filteredimdbdata.head()
```

```python
def reviews(url):
    response_reviews = get(url)
    reviews_soup = BeautifulSoup(response_reviews.text, 'html.parser')
    # Searching all reviews
    reviews_containers = reviews_soup.find_all('div', class_ = 'imdb-user-review')
    # Looping through each review and extracting title and body
    movie_name = []
    review_titles = []
    review_content = []
    reviewer_name = []
    review_date = []
    feedback = []
    movienamecontent = reviews_soup.find(id="main")
    movienamebox = movienamecontent.find('div', class_='subpage_title_block__right-column')
    moviename = movienamebox.find('a').text
    for review_container in reviews_containers:
        movie_name.append(moviename)
        review_title = review_container.find('a', class_ = 'title').text.strip()
        review_body = review_container.find('div', class_ = 'text').text.strip()
        review_titles.append(review_title)
        review_content.append(review_body)
        review_name = review_container.find('span', class_ = 'display-name-link')
        reviewer_name.append(review_name.find("a").text)
        date_of_review = review_container.find('span', class_ = 'review-date').text.strip()
        review_date.append(date_of_review)
        rvfb = review_container.find('div', class_ = 'actions text-muted').text.strip()
        feedback.append(rvfb.split('.')[0])   
    reviews = pd.DataFrame({'movie': movie_name,'reviewer': reviewer_name,'review_date': review_date,'review_title': review_titles,'review_text': review_content,'feedback': feedback})
    return reviews
```

```python
final_reviews = pd.DataFrame()
for url in filteredimdbdata['reviewurls']:
    final_reviews = final_reviews.append(reviews(url))
```
